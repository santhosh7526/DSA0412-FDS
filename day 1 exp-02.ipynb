```python
import pandas as pd

df = pd.read_csv(
    "/content/Sales.csv",
    usecols=["Month_sales", "Price"]
)
```

```python
df["Month_sales"] = df["Month_sales"].astype("category")

```

```python
avg_price_per_month = (
    df.groupby("Month_sales", observed=True)["Price"].mean()
)
```

```python
print(avg_price_per_month)
```
